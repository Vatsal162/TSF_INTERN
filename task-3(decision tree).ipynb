```python
# importing libraries
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
# importing datsset
data=datasets.load_iris()
idata=pd.DataFrame(data.data,columns=data.feature_names)
idata.head()
```

```python
idata.info() # getting some basic idea of the dataset
```

```python
idata['target']=data.target
```

```python
x=idata.drop('target',axis=1)
y=idata['target']
```

```python
# splitting and training the data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
dtree=DecisionTreeClassifier()
dtree.fit(x_train,y_train)
```

```python
# visualising the graph
feat=["sepal length", "sepal-width","petal-length","petal-width"]
classes=["setosa","versicolor","virginica"]
fig,axes=plt.subplots(nrows=1,ncols=1,figsize=(4,4),dpi=300)
tree.plot_tree(dtree,feature_names=feat, class_names=classes,filled=True)
```
